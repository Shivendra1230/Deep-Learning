```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train
```

```python
model = Sequential()

model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(10, activation='softmax'))# for more than 2 classes we use softmax activation function
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])# saprse_categorical_crossentropy is used when the labels are in integer format and categorical_crossentropy is used when the labels are in one hot encoded format where as categorical_crossentropy is used when the labels are in one hot encoded format where as sparse_categorical_crossentropy is used when the labels are in integer format
```

```python
history=model.fit(X_train, y_train, epochs=25, validation_split=0.2)# validation_split is used to split the training data into training and validation data. Here 20% of the training data is used for validation and 80% is used for training. 
```

```python
y_prob=model.predict(X_test)
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['val_accuracy'])
plt.plot(history.history['accuracy'])
```
